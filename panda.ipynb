```python
data = {
    'Company': ['GOOG', 'GOOG', 'MSFT', 'MSFT', 'FB', 'FB'],
    'Person': ['Sam', 'Charlie', 'Amy', 'Vanessa', 'Carl', 'Sarah'],
    'Sales': [200, 120, 340, 124, 243, 350]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
byComp = df.groupby('Company')
```

```python
byComp['Sales'].mean()
```

```python
byComp['Sales'].sum()
```

```python
byComp['Sales'].std()
```

```python
byComp['Sales'].std().loc['FB']
```

```python
byComp.count()
```

```python
byComp.describe().transpose()
```
